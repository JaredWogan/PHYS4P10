#### 2D plotting

Sometimes data is three dimensional, for example $f(x,y)$. In this case there are two independant variables, and the number of data points is $n_x\times n_y$. For every point in space $(x_i,y_i)$, we have a third value to represent. Sometimes we can show this as a 3 dimensional surface, whose height above the $(x,y)$ plane is the value of the function $f(x,y)$. Alternatively, we will introduce the `matplotlib` functions that draw 2 dimensional pictures and color them according to the value of $f$. This is different than a scatter plot, where each data point is just the pair of points $(x_i,y_i)$.

We will need more functions from `matplotlib` to do the plotting, and helper functions for handling the colors. Since we only have data at a single point in space, in order to make a picture the `matplotlib` functions must be able to use additional information and draw a polygon (usually square) around that point, and color the pixels inside the appropriate color. For example, in this figure, there are $9\times9$ data points, but a box must be drawn around each one and colored. Each polygon's size is based on the number of pixels of the final width of the image.

<img src="images/peaks2d.png">

Since `matplotlib` must calculate the polygons around the data to draw the figure, it can also calculate paths through $(x,y)$ and around the data that represent paths of equal values. It can then draw these as contours of a surface like this

<img src="images/contour.png">


````python
%matplotlib inline
#
import numpy
import matplotlib.pyplot as plot
import matplotlib.colors as colors
import matplotlib.ticker as ticker
````

Let's first load the data using `numpy.loadtxt`. 

````python
data=numpy.loadtxt("data/circular.txt",float)
````

#### Example 1

<font color="blue">How many data points are there in the data? What's its dimensionality and size in the each dimension?</font>

The data is $N_x\times N_y$ data points. This does not mean there will be that many pixels in the final image. Instead, if the width of the plotted area is $W$ inches, and the figure is drawn at $P$ pixels per inch, then each data point becomes a polygon of $W/N_x$ inches or $WP/N_x$ image pixels wide.

````python
print(f"There are {data.size} data points in a {data.shape} array.")
````

The first step is to decide how we color the data, by choosing the number of distinct colors and what levels of $f(x,y)$ they represent.

`matplotlib.cm.get_cmap`  Get a colormap instance, defaulting to rc values if name is None.

Please refer to this website for color choices:
https://matplotlib.org/users/colormaps.html

`matplotlib.ticker.MaxNLocator` Finds up to a max number of intervals with ticks at nice locations.

`matplotlib.ticker.MaxNLocator.tick_values` Return the values of the located ticks given `vmin` and `vmax`.

`matplotlib.colors.BoundaryNorm` Generate a colormap index based on discrete intervals.

To actually plot the data, we can create a 2D color plot or colored contour plot.

`matplotlib.pyplot.pcolormesh` Create a pseudocolor plot with a non-regular rectangular grid.

`matplotlib.pyplot.contour` The functions `contour` and `contourf` draw contour lines and filled contours, respectively. 

`matplotlib.pyplot.clabel` Adds labels to line contours in `cs`, where `cs` is a ContourSet object returned by `contour`.

````python
cmap1 = plot.get_cmap('winter') 
cmap2 = plot.get_cmap('viridis')
cmap3 = plot.get_cmap('cool')
levels1 = ticker.MaxNLocator(nbins=20).tick_values(data.min(), data.max())
levels2 = ticker.MaxNLocator(nbins=5).tick_values(data.min(), data.max())
norm1 = colors.BoundaryNorm(levels1, ncolors=cmap1.N, clip=True)

plot.figure(num=None, figsize=(5, 5), dpi=80, facecolor='w', edgecolor='k')
plot.pcolormesh(data,cmap=cmap1,norm=norm1)
plot.colorbar()
plot.show()
plot.close()

plot.figure(num=None, figsize=(5, 5), dpi=80, facecolor='w', edgecolor='k')
CS=plot.contour(data,levels=levels2,cmap=cmap2)
plot.clabel(CS, inline=1, fontsize=8)
plot.show()
plot.close()

plot.figure(num=None, figsize=(5, 5), dpi=80, facecolor='w', edgecolor='k')
plot.imshow(data,cmap=cmap3)
plot.show()
plot.close()
````

#### More about working with data from other sources.

Many scientific instruments allow you to analyse data using software provided by the manufacturer. In most cases, you can find the results you need this way. Sometimes though, you might want to analyse the data in a special way, or you may still want to have the data in a file to prepare your own figures for scientific publication.

The software that collects the data can often export the data as text files. We might use `numpy.loadtxt` in that case. However, sometimes the text file contains non-standard characters, or be unique to Windows or Linux software. I have collected a few data sets from instruments around the Physics department that this is the case.

**Character encoding** is the format used to represent readable characters. Every computer file is nothing but bytes of zeros and ones, so when you look at a text file, the encoding tells the program how many bits to read to translate into a single letter or number.

ASCII uses 7-bits for $2^7=128$ characters. Once you use 52 letters (upper and lowercase), 10 digits, all the punctuation marks typically found on an English keyboard, and a variety of controls like tabs, spaces, and indicating the ends of lines, there is no room in to hold accented or non-English letters.

Thus there are a number of character encodings that expand the amount of characters that can be represented in a text file. For example, [UTF-8][1] is very popular for multi-language web pages.

[1]: https://en.wikipedia.org/wiki/UTF-8

`numpy.loadtxt` may have trouble with anything except ASCII enocding in a file. However, since Python knows most character encodings, we can text read files as *strings*, taking into account their encoding, which eliminates the problem. Then we have to convert the string *back into*  a file-type variable so that `numpy.loadtxt` can work.

<font color="blue"> The file is of type ASCII, which means we could use `numpy.loadtxt`. However, there are 38 lines of addtional *header* information before the data. The data is in two columns.</font>

````python
import numpy
import io
import matplotlib.pyplot as plot
#
# Note that the number of rows of header information is different
# in the txt file that the csv file.
#
filedata_string = open('data/2018-11-22_INDIUM.txt','r',encoding='ASCII').read()
filedata_io = io.StringIO(filedata_string)
data = numpy.loadtxt(filedata_io,unpack=True,skiprows=37)

#
# This requires using the plot.twinx command to switch axis on which to plot.
# After using plot.twinx, all plot.something commands are acting on the new axis.
#
plot.figure(figsize=(9,11))
plot.suptitle(r'Indium differential scanning calorimetry scan', fontsize=16, color='black')
plot.subplot(211)
plot.plot(data[0],'r-',linewidth=2)
plot.ylabel(r'Temp (C)', fontsize=16, color='red')
plot.subplot(212)
plot.plot(data[1],'b-',linewidth=2)
plot.ylabel(r'Heat Flow (mW)', fontsize=16, color='blue')
plot.xlabel(r'Data point', fontsize=16, color='black')
plot.rc('xtick',labelsize=14)
plot.rc('ytick',labelsize=14)
plot.tight_layout()
plot.show()
plot.close()
````

#### Example 3

The next set of data are from the KSV/Nima Langmuir trough in H208, and are compression isotherms of a single-molecule thick layer of lipid molecules on the surface of water. 

     12-15-2018_POPC.csv
     
I had difficulty opening this data with `numpy.loadtxt`, but numpy has more robust methods as well, like `numpy.genfromtext`.

* What's the encoding of this file? 
* How many columns of data are there? 
* How many lines of additional information are there before the actual data begins?

Follow the method above to read the data, this time using `numpy.genfromtext`. Here is one example that loads the data.
```python
filedata_string = open('12-15-2018_POPC.csv','r',encoding='ISO-8859-1').read()
filedata_io = io.StringIO(filedata_string)
data = numpy.genfromtxt(filedata_io, dtype=float, delimiter=',', skip_header=1, unpack=True)
```

There is a lot of data in this file. First, it has multiple columns of data, and it extends well past the interesting parts. This time, I will use the `numpy.unpack` function to extract the data below.

Show the isotherm by plotting surface pressure ($\pi$ or variable P1) versus mean molecular area (surface area divided by the number of molecules, $A/N$ or variable Mma), in other words a plot of the ideal gas law for 2D ($\pi A=Nk_BT$) for constant $T$. You will note that over time the surface pressure settles to a vlaue around 30 mN/m for POPC and 45 mN/m for DPPC. So only plot data up to time point of ~1400 seconds.

````python
import numpy
import io
import matplotlib.pyplot as plot

filedata_string = open('data/12-15-2018_POPC.csv','r',encoding='ISO-8859-1').read()
filedata_io = io.StringIO(filedata_string)
data = numpy.genfromtxt(filedata_io, dtype=float, delimiter=',', skip_header=1, unpack=True)

mma = numpy.extract(data[0]<1400,data[4]) # THE MMA IS THE 5TH COLUMNS OF DATA
pi  = numpy.extract(data[0]<1400,data[5]) # THE SURFACE PRESSURE IS THE 6TH COLUMN OF DATA
                    
plot.figure(figsize=(9,7))    
plot.plot(mma,pi,'r-',linewidth=4)
plot.text(170, 20, r'Unclear phases?', fontsize=16)
plot.xlabel(r'Mean molecular area ($\AA^2$)', fontsize=16, color='black')
plot.ylabel(r'Surface Pressure (mN/m)', fontsize=16, color='black')
plot.title(r'POPC monlayer isotherm', fontsize=16, color='black')
plot.show()
plot.close()
````

#### Example 4

The next is raw data from the Rigaku X-ray machine in H305;     

`05-17-2018_Cr2ALC.ras`

The three data columns are scattering angle $2\theta$, x-ray intensity, and counting time. You will see there is *a lot* of additoinal information in the *header* of the file before the data begins. 

Here I had problems making the `io.StringIO` to `numpy.genfromtext` method work. That is because although the `ras` data format is text, it had a few Japanese characters in it and was encoded as `Shift_JIS` format. Even though Python can read this encoding, `numpy.genfromtext` still had problems.  When I deleted the Japanese characters, it worked. 

Also, rather than skipping lines of header informaion, it was clear that all lines in the file that had no data started with an asterisk `*`. In this case, we can tell `numpy.loadtxt` to skip lines starting with `*` with the `comments` option.
     
* Plot the Cr2ALC data set in its full scan range of $2\theta$.

````python
import numpy
import io
import matplotlib.pyplot as plot

filedata_string = open('data/05-17-2018_Cr2ALC.ras','r',encoding='ISO-8859-1').read()
filedata_io = io.StringIO(filedata_string)
data = numpy.loadtxt(filedata_io,comments="*",unpack=True,dtype=float)
                         
plot.figure(figsize=(12,7))    
plot.plot(data[0],data[1],'r-',linewidth=1)
plot.xlabel(r'Scattering angle $2\theta$ (deg)', fontsize=16, color='black')
plot.ylabel(r'Intensity (counts)', fontsize=16, color='black')
plot.title(r'X-ray powder diffraction of Cr2ALC', fontsize=16, color='black')
plot.show()
plot.close()
````
