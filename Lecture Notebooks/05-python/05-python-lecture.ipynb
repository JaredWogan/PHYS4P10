$$\frac{1}{\lambda}  = R\left(\frac{1}{m^2} -\frac{1}{n^2}\right)$$

```python
# CALCULATE RYDBERG WAVELENGTHS
R = 1.097E-2 # UNITS OF 1/nm

for m in [1,2,3] :
    print("Spectral lines for m: ", m)
    for k in [1,2,3,4] :
        n = m + k
        invlambda =  R * ((1/m**2) - (1/n**2))
        #print("    n:",n, "is", 1/invlambda, " nm.")
        print(f"    n: {n:d} is {1/invlambda:1.3f} nm")
```

```python
#version 0.1
def my_factorial(n):
    f = 1
    for k in range(1,n+1) :
        f *= k
    # END OF FOR LOOP
    return f
# END OF MY_FACTORIAL FUNCTION 
```

```python
g = my_factorial(6)
print(g)
```

```python
g = my_factorial(100)
print(g)
```

```python
# CALCULATE LENGTH OF A VECTOR GIVEN IN CYLINDRICAL COORDINATES
def distance(r,theta,z):
    x = r * numpy.cos(theta)
    y = r * numpy.sin(theta)
    d = numpy.sqrt(x**2+y**2+z**2)
    return d
```

```python
d = distance(2.0,0.1,-1.5)
print(d)
```

```python
#FIND ALL FACTOS OF INTEGER n
def factors(n) : 
    factorlist = []
    k = 2
    while k <= n :
        if (n%k == 0) :
            factorlist.append(k)
            n = n//k
        k += 1
    return factorlist
```

```python
f = factors(10_000_000)
print(f)
```

LOGISTIC MAP

$$x_{n-1}=rx_{n}(1-x_{n})$$
$$x_{n}=\left[0,1\right)$$
$$r\sim(0,4)$$

If I start at some $x$ value, what does $x_{n}$ become when $n$ is large?

```python
%matplotlib inline
import matplotlib.pyplot as plot

r_values = numpy.arange(2.8,5.5,0.01,dtype=float)

x_plot_data = numpy.zeros([100])
r_plot_data = numpy.zeros([100])


plot.figure(figsize=(9,6))
for r in r_values:
    # ----
    x = 0.5
    for i in range(1500) :
        xp = r * x * (1-x)
        x = xp
    # -----
    
    for i in range(100) :
        xp = r * x * (1-x)
        x = xp
        x_plot_data[i] = x
        r_plot_data[i] = r
    # -----
    
    plot.scatter(r_plot_data,x_plot_data, color='k', marker=".")
    
plot.show()
```
